```python
df1 = df[df.Oldpeak<=(df.Oldpeak.mean()+3*df.Oldpeak.std())]
```

```python
df1.shape
```

```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df2 = df1[df1.RestingBP<=(df1.RestingBP.mean()+3*df1.RestingBP.std())]
```

```python
from sklearn.preprocessing import LabelEncoder
for column in df2.columns:
    
    if df2[column].dtypes == np.number:
        continue
    df2[column] = LabelEncoder().fit_transform(df2[column])
```

```python
df2.head()
```

```python
x = df2.drop(['HeartDisease'],axis=1)
```

```python
x
```

```python
y = df2['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
xscaled = scaler.fit_transform(x)
xscaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrian,ytest = train_test_split(xscaled,y,test_size=.25,random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(),x,y,cv =5)
scores.mean()
```

# use bagging now with svm

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(base_estimator=SVC(), n_estimators=100, max_samples=0.8, random_state=0)
scores = cross_val_score(bag_model, x,y, cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
scores = cross_val_score(DecisionTreeClassifier(random_state=42),x,y,cv=5)
scores.mean()
```

```python
bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(random_state=0), 
    n_estimators=100, 
    max_samples=0.9, 
    oob_score=True,
    random_state=0
)

scores = cross_val_score(bag_model, x, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
scores = cross_val_score(RandomForestClassifier(),x,y,cv = 5)
scores.mean()
```
